```python
import pandas as pd
import sqlite3 as sql
```

```python
filepath = "./reported_hospital_capacity_admissions_facility_level_weekly_average_timeseries_20210221.csv"
```

```python
df = pd.read_csv(filepath)
```

```python
df = df[['collection_week', 'state', 'hospital_name', 'city', 'zip', 'staffed_adult_icu_bed_occupancy_7_day_avg', 'total_staffed_adult_icu_beds_7_day_avg']]
```

```python
df['zip'] = df['zip'].fillna(-1)
df['zip'] = df['zip'].astype(int).astype(str)
```

```python
df = df[df['collection_week'] == "2021-02-12"]
df['staffed_adult_icu_bed_occupancy_7_day_avg'] = df['staffed_adult_icu_bed_occupancy_7_day_avg'].replace(-999999.0, 'NaN')
df['total_staffed_adult_icu_beds_7_day_avg'] = df['total_staffed_adult_icu_beds_7_day_avg'].replace(-999999.0, 'NaN')
df['open_adult_icu_beds'] = df['total_staffed_adult_icu_beds_7_day_avg'] - df['staffed_adult_icu_bed_occupancy_7_day_avg']
```

```python
df
```

```python
def s(raw):
    return str(raw).replace("'", "''")
```

```python
with sql.connect("database.db") as con:
    cur = con.cursor()
    cur.execute("""CREATE TABLE occupancy 
        (date text, 
            state text,
            hospital text,
            city text,
            zip text,
            open_adult_icu_beds int
        )
    """)
```

```python
count = 0

with sql.connect("database.db") as con:
    cur = con.cursor()
    for index, row in df.iterrows():
        try:
            if (row.staffed_adult_icu_bed_occupancy_7_day_avg == -999999) or (row.total_staffed_adult_icu_beds_7_day_avg == -999999):
                cnxnstr = f"INSERT INTO occupancy (date, state, hospital, city, zip, open_adult_icu_beds) values ('{row.collection_week}', '{row.state}', '{s(row.hospital_name)}', '{s(row.city)}', '{row.zip}', NULL)"
            else:
                cnxnstr = f"INSERT INTO occupancy (date, state, hospital, city, zip, open_adult_icu_beds) values ('{row.collection_week}', '{row.state}', '{s(row.hospital_name)}', '{s(row.city)}', '{row.zip}', {int(row.total_staffed_adult_icu_beds_7_day_avg - row.staffed_adult_icu_bed_occupancy_7_day_avg)})"
            cur.execute(cnxnstr)
        except:
            print(cnxnstr)
    con.commit()
```
